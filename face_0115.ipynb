```
import glob
import face_recognition
import os
import pandas as pd
names = []
encodings = []
for fp in glob.glob("/gdrive/My Drive/face0115/*"):
    img = face_recognition.load_image_file(fp)
    en = face_recognition.face_encodings(img)[0]
    encodings.append(list(en))
    n = os.path.split(fp)[-1].split(".")[0]
    names.append(n)
df = pd.DataFrame({
    "Name":names,
    "Encoding":encodings
})
df.to_csv("/gdrive/My Drive/face0115.csv",
          encoding="utf-8",
          index=False)
df
```

```
import json
names = list(df["Name"])
encodings = list(map(lambda x:json.loads(x), df["Encoding"]))
print(names)
print(encodings)
```

```
import requests
import face_recognition
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
url = input("請輸入圖片網址:")
response = requests.get(url, stream=True, verify=False)
img = face_recognition.load_image_file(response.raw)
# 如果你追求速度, 就不用改這, 如果追求精準度, 就使用cnn做人臉定位
locs = face_recognition.face_locations(img, model="cnn")
ens = face_recognition.face_encodings(img, known_face_locations=locs)
for i, (en, loc) in enumerate(zip(ens, locs)):
    print("找到臉:", loc)
    distances = face_recognition.face_distance(en, encodings)
    for n, d in sorted(zip(names, distances), key=lambda x:x[1], reverse=True):
        print("跟", n, "的距離:", d)
    mind = min(distances)
    idx = distances.argmin()
    if mind < 0.5:
        print("這張臉就是:", names[idx])
    else:
        print("沒有辨識到, 但是比較像:", names[idx])
    print("-" * 30)
plt.axis("off")
for top, right, bottom, left in locs:
    cv2.rectangle(img, (left, top), (right, bottom), (255, 255, 255), 10)
plt.imshow(img)
```
