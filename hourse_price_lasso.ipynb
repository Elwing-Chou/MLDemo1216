```python
train = train_df.loc[:,"MSSubClass":"SaleCondition"]
test = test_df.loc[:,"MSSubClass":"SaleCondition"]
x = pd.concat([train, test])
y_train = train_df["SalePrice"]
x
```

```python
# 先把數字類型的欄位存起來!
numeric = (x.columns[x.dtypes != "object"]
            .drop(["MSSubClass"]))
numeric
```

```python
x = pd.get_dummies(x)
x = pd.get_dummies(x, columns=["MSSubClass"])
```

```python
s = x.isna().sum()
s[s > 0]
```

```python
from scipy.stats import skew
# dataframe的apply拿到的東西是一個一個Series
sk = x[numeric].apply(lambda s:skew(s.dropna()))
sk[sk > 3].sort_values(ascending=False)
```

```python
from scipy.special import boxcox1p
# 做完boxcox1p轉換
need_saved = sk[sk > 5]
x_skew = boxcox1p(x[need_saved.index], 0.15)
# 觀察一下skewness
sk = x_skew.apply(lambda s:skew(s.dropna()))
sk.sort_values(ascending=False)
```

```python
# 真的把所有東西換成常態分佈
x[need_saved.index] = x_skew
# x = x.drop(["PoolArea"], axis=1)
y = np.log1p(y_train)
x
```

```python
# numeric = numeric.drop(['PoolArea'])
x[numeric] = x[numeric].fillna(x[numeric].mean())
s = x.isna().sum()
s[s > 0]
```

```python
x_train = x.iloc[:train_df.shape[0]]
x_test = x.iloc[train_df.shape[0]:]
x_test
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=0.0004)
lasso.fit(x_train, y)
```

```python
pre = lasso.predict(x_test)
pre
```

```python
pre_exp = np.expm1(pre)
upload = pd.DataFrame({
    "Id":test_df["Id"],
    "SalePrice":pre_exp
})
upload.to_csv("house_lasso.csv", encoding="utf-8", index=False)
upload
```
