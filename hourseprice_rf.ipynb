```python
import pandas as pd
train_df = pd.read_csv("house/train.csv")
test_df = pd.read_csv("house/test.csv")
train_df
```

```python
train = train_df.loc[:,"MSSubClass":"SaleCondition"]
test = test_df.loc[:,"MSSubClass":"SaleCondition"]
x = pd.concat([train, test])
y_train = train_df["SalePrice"]
x
```

```python
# 先把數字類型的欄位存起來!
numeric = (x.columns[x.dtypes != "object"]
            .drop(["MSSubClass"]))
numeric
```

```python
# 存下原本欄位, 讓你看一下新增多少欄位
ori_columns = x.columns
```

```python
x = pd.get_dummies(x)
x = pd.get_dummies(x, columns=["MSSubClass"])
x
```

```python
# 被做One-hot的欄位
set(ori_columns) - set(x.columns)
```

```python
# 被做成什麼欄位
set(x.columns) - set(ori_columns)
```

```python
x[numeric].isna().sum()
```

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.impute.IterativeImputer.html
# all_data = all_data.fillna(all_data.median())
import numpy as np
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
imp = IterativeImputer(max_iter=10, random_state=0)
x_impute = imp.fit_transform(x) 
pd.DataFrame(x_impute).isna().sum()
```

```python
x_impute_df = pd.DataFrame(x_impute)
x_train = x_impute_df.iloc[:train_df.shape[0]]
x_test = x_impute_df.iloc[train_df.shape[0]:]
x_test
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
reg = RandomForestRegressor()
p = {
    "n_estimators":range(10, 100, 5),
    "max_depth":range(5, 10)
}
grid = GridSearchCV(reg, p, scoring="r2", cv=10)
grid.fit(x_train, y_train)
print("最好分數:", grid.best_score_)
print("最好參數:", grid.best_params_)
```

```python
reg = RandomForestRegressor(max_depth = 8, n_estimators = 75)
reg.fit(x_train, y_train)
pre = reg.predict(x_test)
pre
```

```python
upload = pd.DataFrame({
    "Id":test_df["Id"],
    "SalePrice":pre
})
upload.to_csv("house_rf.csv", encoding="utf-8", index=False)
upload
```
